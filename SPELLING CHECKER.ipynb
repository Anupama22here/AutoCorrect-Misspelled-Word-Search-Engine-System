## Autocorrection


1. Split Operation
[('', 'hello'), ('h', 'ello'), ('he', 'llo'), ('hel', 'lo'), ('hell', 'o'), ('hello', '')]

```python
def split(word):
    a = [ (word[:i], word[i:]) for i in range(len(word) + 1)]
    return a
print(split('forest'))
```

2. Delete Operation

```python
def delete(word):
    b=[left+right[1:] for left,right in split(word) if right]
    return b
print(delete('forest'))
```

3. Swap Operation

```python
def swap(word):
    c=[left+right[1]+right[0]+right[2:] for left,right in split(word) if len(right)>1 ]
    return c
print(swap('forest'))
```

4. Replace Operation

```python
def replace(word): 
    d=[left + center + right[1:] for left, right in split(word) if right for center in string.ascii_lowercase]
    return d
print(replace('forest'))
```

5. Insert Operation

```python
def insert(word):
    e=[left + center + right[1:] for left, right in split(word) for center in string.ascii_lowercase]
    return e
print(replace('forest'))
```

```python
def level_one_edits(word):
    a=set((delete(word) + swap(word) + replace(word) + insert(word)))
    return a
print(level_one_edits('love'))
```

```python
def level_two_edits(word):
    b=set(e2  for e1 in level_one_edits(word) for e2 in level_one_edits(e1))
    return b
print(level_two_edits('candy'))
```
